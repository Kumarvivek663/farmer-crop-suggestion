```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns  
```

```python
df = pd.read_csv('farmer.csv')
df.head(40)
```

```python
df['label'].value_counts()
```

```python
group = df.groupby('label')
```

```python
group[['N']].mean()
```

```python
temperature = group['temperature'].agg(min_temperature='min',max_temperature='max',avg_temperature='mean')
```

```python
temperature 
```

```python
temperature.reset_index(inplace=True)
```

```python
sns.catplot(x='label',y='avg_temperature',data=temperature,kind='bar',ec='k',color='blue')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.catplot(x='label',y='min_temperature',data=temperature,kind='bar',ec='k',color='blue')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.catplot(x='label',y='max_temperature',data=temperature,kind='bar',ec='k',color='blue')
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scaler = StandardScaler()
x_transform = std_scaler.fit_transform(x)
```

```python
x.transform
```

```python
pd.DataFrame(x_transform)
```

```python
sns.kdeplot(pd.DataFrame(x_transform)[3])
plt.show()
```

```python
sns.kdeplot(x['temperature'])
plt.show()
```

```python
from sklearn.cluster import KMeans  
kmeans_obj = KMeans()  
labels = kmeans_obj.fit_predict(x_transform)    #algorithm  trained and prediction
```

```python
set(labels)
```

```python
df['cluster_no.'] = labels
```

```python
import os,joblib 
os.makedirs('models',exist_ok=True) 
joblib.dump(std_scaler,'./models/std_scaler.lb') 
joblib.dump(kmeans_obj,'./models/kmeans_model.lb')
```
